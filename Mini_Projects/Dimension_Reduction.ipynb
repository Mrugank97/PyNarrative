### **Project Goals**

- Evaluate the trade-offs between dimensionality reduction and clustering accuracy.

```python
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import time

wine=load_wine()
x1=wine.data
poke=StandardScaler()
x=poke.fit_transform(x1)

start1=time.time()
kmeans1=KMeans(n_clusters=3,random_state=42,n_init=10)
kmeans1.fit(x)
end1=time.time()

print("## K-means on Original dataset ##")
print("Number of clusters = 3")
print("Time required:", (end1-start1))
print("Size of the dataset wine:", x.shape,"\n")

pca1=PCA(n_components=2)
x_pca1=pca1.fit_transform(x)

start_2=time.time()
kmean2=KMeans(n_clusters=3,random_state=42,n_init=10)
kmean2.fit(x_pca1)
end_2=time.time()

label=kmean2.labels_
centroid=kmean2.cluster_centers_


plt.subplot(1,2,1)
plt.scatter(x_pca1[:,0],x_pca1[:,1],c=label)
plt.scatter(centroid[:,0],centroid[:,1],marker='*', s=100,c='red', label='centroid')
plt.title('K-means on Wine Dataset with Dim 2')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()

print("## K-means on Wine Dataset with Dimension 2 ##")
print("Number of clusters = 3")
print("Time required for reduced dataset:",end_2-start_2)
print("Size of the reduced dataset:",x_pca1.shape)
print("Explained variance ratio:", pca1.explained_variance_ratio_,"\n")

pca2=PCA(n_components=3)
x_pca2=pca2.fit_transform(x)
start_3=time.time()
kmean3=KMeans(n_clusters=3,random_state=42,n_init=10)
kmean3.fit(x_pca2)
end_3=time.time()

label2=kmean3.labels_
centroid2=kmean3.cluster_centers_

plt.subplot(1,2,2)
plt.scatter(x_pca2[:,0],x_pca2[:,1],c=label2)
plt.scatter(centroid2[:,0],centroid2[:,1],marker='*',s=100,c='red',label='centroid')
plt.title('K-means on Wine Dataset with Dim 3')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()

print("## K-means on Wine Dataset with Dimension 3 ##")
print("Number of clusters = 3")
print("Time required for reduced dataset:",end_3-start_3)
print("Size of the reduced dataset:",x_pca2.shape)
print("Explained variance ratio:", pca2.explained_variance_ratio_,"\n")
```
